# Web Scraping amb Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Configuración del navegador
driver = webdriver.Chrome() 

# Abrir la pagina web
driver.get("http://quotes.toscrape.com")

quotes = driver.find_elements(By.CLASS_NAME, "text") 
authors = driver.find_elements(By.CLASS_NAME, "author")

# Crear una lista
data = []

for quote, author in zip(quotes, authors):
    data.append({'Quote': quote.text.strip(),'Author': author.text.strip()})

driver.quit()

quotes_df = pd.DataFrame(data)

quotes_df
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

# URL de la página
url = "https://es.wikipedia.org/wiki/Premios_Goya"

# Inicializar el navegador
driver = webdriver.Chrome()

# Abrir la página web
driver.get(url)

# Encontrar todas las tablas en la página
tables = driver.find_elements(By.CLASS_NAME, 'wikitable')

# Seleccionar la tabla teniendo en cuenta la web
awards_table = tables[1]

# Encontrar todas las filas de la tabla
rows = awards_table.find_elements(By.TAG_NAME, "tr")

# Crear una lista para almacenar los datos de la tabla
table_data = []

# Iterar sobre cada fila
for i, row in enumerate(rows):
    # Encontrar todas las celdas en la fila
    if i == 0:  # Si es la primera fila (encabezados)
        # Extraer los nombres de las columnas y agregarlos a la lista de datos de la tabla
        headers = row.find_elements(By.TAG_NAME, "th")
        header_data = [header.text for header in headers]
        table_data.append(header_data)
    else:
        cells = row.find_elements(By.TAG_NAME, "td")
        # Extraer el texto de cada celda y agregarlo a la lista de datos de la tabla
        row_data = [cell.text.replace('\n', '') for cell in cells]
        table_data.append(row_data)

# Crear un DataFrame de Pandas con los datos de la tabla y los nombres de las columnas
df = pd.DataFrame(table_data[1:], columns=table_data[0])

# Cerrar el navegador
driver.quit()

df
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# URL de la página inicial
url = "https://www.premiosgoya.com/"

# Inicializar el navegador
driver = webdriver.Chrome()

# Lista para almacenar los títulos de todas las páginas
titulos_totales = []

# Recopilar títulos de todas las páginas
while True:
    driver.get(url)
    encabezados = driver.find_elements(By.XPATH, '//h2[@class="relacionado__titulo"]')
    titulos_pagina = [encabezado.text for encabezado in encabezados]
    titulos_totales.extend(titulos_pagina)
    
    try:
        # Hacer clic en el enlace "Más noticias" si está presente
        enlace_mas_noticias = driver.find_element(By.XPATH, '//a[@class="enlace-gordo"]')
        url = enlace_mas_noticias.get_attribute('href')
    except:
        # Si no hay más enlaces, salir del bucle
        break

driver.quit()
```

```python
print(titulos_totales)
```

```python
# Crear DataFrame con los títulos recopilados
df_goya = pd.DataFrame({'Titulos': titulos_totales})

df_goya
```
